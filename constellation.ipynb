Constellation

Three characters { #, *, . } represents a constellation of stars and galaxies in space. Each galaxy is demarcated by # characters. There can be one or many stars in a given galaxy. Stars can only be in the shape of vowels { A, E, I, O, U }. A collection of * in the shape of the vowels is a star. A star is contained in a 3x3 block. Stars cannot be overlapping. The dot(.) character denotes empty space.

Given 3xN matrix comprising of { #, *, . } character, find the galaxy and stars within them.

Note: Please pay attention to how vowel A is denoted in a 3x3 block in the examples section below.

Constraints

3 <= N <= 10^5

Input

Input consists of a single integer N denoting the number of columns.

Output

The output contains vowels (stars) in order of their occurrence within the given galaxy. The galaxy itself is represented by the # character.

Example 1

Input

18

* . * # * * * # * * * # * * * . * .

* . * # * . * # . * . # * * * * * *

* * * # * * * # * * * # * * * * . *

Output

U#O#I#EA

Explanation

As can be seen, the stars make the image of the alphabets U, O, I, E, and A respectively.

Example 2

Input

12

* . * # . * * * # . * .

* . * # . . * . # * * *

* * * # . * * * # * . *

Output

U#I#A

Explanation

As it can be seen the stars make the image of the alphabet U, I, and A.

```python
# vowels possible : A E I O U 
# 3 X 3 matrix will be given 
# Input = single Integer N denoting the number of columns 
def constellation(user_input: list[list[str]], n:int) -> str :
    rep_dict = {
        '.*.****.*' : 'A',
        '*********' : 'E',
        '***.*.***' : 'I',
        '****.****' : 'O',  
        '*.**.****' : 'U'}
    ans = str()
    i = 0
    while i <= n-3:
        temp = user_input[0][0][i:i+3] + user_input[1][0][i:i+3] + user_input[2][0][i:i+3]
        if temp in rep_dict:
            ans += rep_dict.get(temp)
            if i+3 < n-2:
                if user_input[0][0][i+3] == '#':
                    ans += '#'
                    i += 4
                else :
                    i += 3
            else : break
            
        else :
            i+=1
    return ans



def main() -> None:
    n = int(input('Enter the number of columns'))    # here n represents the number of columns
    user_input = [[''] for _ in range(3)]
    for lst in user_input :
        # for _ in range(n):  # accroding to the question
        lst[0] = ''.join(input().split(' '))
    found = constellation(user_input, n)
    print(found)


if __name__ == '__main__':
    main()
```
